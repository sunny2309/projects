```python
def get_ohlc_data(ticker, start_date, end_date, freq="DAILY"):
    '''
    freq = [MONTHLY, WEEKLY, DAILY, EVERY_MINUTE, EVERY_FIVE_MINUTES, EVERY_FIVE_MINUTES, 
            EVERY_FIFTEEN_MINUTES, EVERY_THIRTY_MINUTES]
    '''
    period_type = client.Client.PriceHistory.PeriodType.YEAR if freq in ["MONTHLY", "WEEKLY","DAILY"] else client.Client.PriceHistory.PeriodType.DAY
    
    freq_type_map = {"DAILY": client.Client.PriceHistory.FrequencyType.DAILY,
                     "WEEKLY": client.Client.PriceHistory.FrequencyType.WEEKLY, 
                     "MONTHLY": client.Client.PriceHistory.FrequencyType.MONTHLY}
    
    frequency_type = freq_type_map.get(freq, client.Client.PriceHistory.FrequencyType.MINUTE) 
    
    if freq in ["MONTHLY", "WEEKLY", "DAILY"]:
        r = c.get_price_history(ticker,
            period_type=period_type,
            frequency_type=frequency_type,
            start_datetime = datetime.strptime(start_date, "%Y%m%d"),
            end_datetime=datetime.strptime(end_date, "%Y%m%d"),
            need_extended_hours_data=False
            )
    else:
        freq_map = {"EVERY_MINUTE": client.Client.PriceHistory.Frequency.EVERY_MINUTE, 
                    "EVERY_FIVE_MINUTES": client.Client.PriceHistory.Frequency.EVERY_FIVE_MINUTES,
                    "EVERY_TEN_MINUTES": client.Client.PriceHistory.Frequency.EVERY_TEN_MINUTES,
                    "EVERY_FIFTEEN_MINUTES": client.Client.PriceHistory.Frequency.EVERY_FIFTEEN_MINUTES,
                    "EVERY_THIRTY_MINUTES": client.Client.PriceHistory.Frequency.EVERY_THIRTY_MINUTES}
        
        r = c.get_price_history(ticker,
            period_type=period_type,
            #period=client.Client.PriceHistory.Period.THREE_DAYS,
            frequency_type=client.Client.PriceHistory.FrequencyType.MINUTE,
            frequency=freq_map[freq],
            start_datetime = datetime.strptime(start_date, "%Y%m%d"),
            end_datetime=datetime.strptime(end_date, "%Y%m%d"),
            need_extended_hours_data=False
            )

    data = r.json()
    try:
        df = pd.DataFrame(data["candles"])
        df.datetime = pd.to_datetime(df.datetime, unit="ms")
        df = df.sort_values(by=["datetime"])
        df = df[df.datetime >= datetime.strptime(start_date, "%Y%m%d")]
        df = df.set_index("datetime")
    except Exception as e:
        df = None
    return {ticker : df}
```

```python
def get_ohlc_data_for_tickers(tickers, start_date, end_date, freq="DAILY"):
    final_df = None
    for ticker in tickers:
        try:
            data = get_ohlc_data(ticker, start_date, end_date, freq)
            df = data[ticker]
            df = df[["close"]].rename(columns={"close":ticker})
            if isinstance(final_df, pd.DataFrame):
                final_df = final_df.join(df)
            else:
                final_df = df.copy()
        except Exception as e:
            print(e)
            print("Failed Ticker : ", ticker)
            
    return final_df
```

```python
def get_ohlc_data_for_tickers_parallel(executor, tickers, start_date, end_date, freq="DAILY"):
    futures = []
    for ticker in tickers:
        future = executor.submit(get_ohlc_data, ticker, start_date, end_date, freq)
        futures.append(future)
        
    final_df = None
    for future in as_completed(futures):
        try:
            data = future.result()
            ticker = list(data.keys())[0]
            df = data[ticker]
            df = df[["close"]].rename(columns={"close":ticker})
            if isinstance(final_df, pd.DataFrame):
                final_df = final_df.join(df)
            else:
                final_df = df.copy()
        except Exception as e:
            print(e)
            #print("Failed Tickers : ", ticker)
    return final_df
```

```python
%%time

tpe = ThreadPoolExecutor(max_workers=50)

get_ohlc_data_for_tickers_parallel(tpe, tickers,"20201025", "20201031", "DAILY")
```

```python
%%time

ppe = ProcessPoolExecutor(max_workers=50)

get_ohlc_data_for_tickers_parallel(ppe, tickers,"20201025", "20201031", "DAILY")
```
