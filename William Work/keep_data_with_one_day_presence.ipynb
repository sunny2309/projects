```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/home/sunny/Downloads/test_portfolio_out.csv", index_col=0, parse_dates=True)
df.head()
```

```python
df.tail()
```

```python
data = df.values

modified_data = []
for i in range(data.shape[0]):
    modified_row = []
    for j in range(data.shape[1]):
        if i==0:
            modified_row.append(data[i][j] if (~np.isnan(data[i][j])) and np.isnan(data[i+1][j]) else np.nan)
        elif i == data.shape[0]-1:
            modified_row.append(data[i][j] if (~np.isnan(data[i][j])) and np.isnan(data[i-1][j]) else np.nan)
        else:
            modified_row.append(data[i][j] if (~np.isnan(data[i][j])) and np.isnan(data[i-1][j]) and np.isnan(data[i+1][j]) else np.nan)
    
    modified_data.append(modified_row)

modified_data = np.array(modified_data)
```

```python
modified_data.shape, data.shape
```

```python
final_df = pd.DataFrame(data=modified_data, index=df.index, columns=df.columns)
final_df.head()
```

```python
final_df.to_csv("modified.csv")
```
